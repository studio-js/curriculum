# Python 기초: 자료형, 조건문, 반복문

이 노트북에서는 Python의 핵심 기초 문법을 다룹니다.  
**int · float · str · bool · list · dict** 자료형과 **if · for · while** 제어 흐름을 익히고,  
리스트 컴프리헨션으로 코드를 간결하게 작성하는 패턴을 실습합니다.

## 1. 기본 자료형 (Data Types)

| 자료형 | 예시 | 설명 |
|--------|------|------|
| `int` | `42` | 정수 |
| `float` | `3.14` | 실수 |
| `str` | `'hello'` | 문자열 |
| `bool` | `True` | 불리언 |
| `list` | `[1, 2, 3]` | 리스트 |
| `dict` | `{'key': 'val'}` | 딕셔너리 |

````python
age = 42
height = 3.14
name = '홍길동'
is_student = True

print('정수:', age, type(age))
print('실수:', height, type(height))
print('문자열:', name, type(name))
print('불리언:', is_student, type(is_student))
````

````python
scores = [85, 92, 78, 96, 88]
print('리스트:', scores)
print('첫 번째:', scores[0])
print('마지막:', scores[-1])
print('슬라이싱 [1:3]:', scores[1:3])
````

````python
person = {'name': '홍길동', 'age': 28, 'job': '데이터 과학자'}
print('이름:', person['name'])
print('나이:', person['age'])
print('직업:', person['job'])
````

## 2. 조건문 (if / elif / else)

조건에 따라 다른 코드를 실행합니다.

```python
if 조건1:
    ...
elif 조건2:
    ...
else:
    ...
```

````python
score = 85

if score >= 90:
    grade = 'A학점'
elif score >= 80:
    grade = 'B학점'
elif score >= 70:
    grade = 'C학점'
elif score >= 60:
    grade = 'D학점'
else:
    grade = 'F학점'

print(f'점수: {score} → {grade}')
````

## 3. 반복문 (for / while)

- **for**: 시퀀스를 순회할 때
- **while**: 조건이 참인 동안 반복

````python
fruits = ['사과', '바나나', '체리']
for idx, fruit in enumerate(fruits):
    print(f'{idx}번: {fruit}')

scores = [88, 92, 85, 95, 90]
total = sum(scores)
print(f'총합: {total}, 평균: {total / len(scores)}')
````

````python
count = 1
while count <= 5:
    print(f'카운트: {count}')
    count += 1
print('종료!')
````

## 4. 리스트 컴프리헨션

```python
[표현식 for 변수 in 시퀀스 if 조건]
```

````python
squares = [x ** 2 for x in range(10)]
print('제곱:', squares)

evens = [x for x in range(10) if x % 2 == 0]
print('짝수:', evens)

scores = [85, 42, 92, 58, 96, 88]
passing = [s for s in scores if s >= 60]
print('합격 점수:', passing)
````

## 5. 종합 실습 — 학생 성적 분석기

자료형, 조건문, 반복문, 컴프리헨션을 모두 활용합니다.

````python
students = [
    {'name': '김민준', 'score': 92},
    {'name': '이서연', 'score': 78},
    {'name': '박지호', 'score': 55},
    {'name': '최수아', 'score': 88},
    {'name': '정태양', 'score': 67},
]

def get_grade(score):
    if score >= 90:   return 'A학점'
    elif score >= 80: return 'B학점'
    elif score >= 70: return 'C학점'
    elif score >= 60: return 'D학점'
    else:             return 'F학점'

print('===== 학생 성적 분석 =====\n')
for s in students:
    grade  = get_grade(s['score'])
    passed = '✅' if s['score'] >= 60 else '❌'
    print(f"{s['name']}: {s['score']}점 → {grade} {passed}")

all_scores = [s['score'] for s in students]
passing    = [s for s in all_scores if s >= 60]

print(f'\n📊 통계')
print(f'  전체 인원: {len(students)}명')
print(f'  평균 점수: {sum(all_scores)/len(all_scores)}점')
print(f'  최고 점수: {max(all_scores)}점')
print(f'  최저 점수: {min(all_scores)}점')
print(f'  합격 인원: {len(passing)}명 ({len(passing)/len(students)*100}%)')
````
